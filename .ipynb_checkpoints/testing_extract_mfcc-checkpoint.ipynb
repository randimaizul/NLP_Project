```python
import librosa 
import numpy as np
from dtw import dtw
import numpy as np
from matplotlib import pyplot as plt
```

```python

with open('dataset/apel/wavToTag.txt') as f:
    labels = array([l.replace('\n', '') for l in f.readlines()])

len(labels)
```

```python
mfccs = {}
for i in range(len(labels)):
    y, sr = librosa.load('dataset/apel/{}.wav'.format(i))
    mfcc = librosa.feature.mfcc(y, sr, n_mfcc=13)
    mfccs[i] = mfcc.T
```

```python
def generate_train_test_set(spr,labels):
    train = []
    test = []
    
    for s in set(labels):
        all = find(labels == s)
        shuffle(all)
        train += all[:-spr].tolist()
        test += all[-spr:].tolist()
    return train, test

def cross_validation(train, test):
    score = 0.0

    for i in test:
        x = mfccs[i]

        dmin, jmin = inf, -1
        for j in train:
            y = mfccs[j]
        
            d = D[i, j]
            if d == -1:
                d, _, _, _ = dtw(x, y, dist=lambda x, y: norm(x - y, ord=1))
                D[i, j] = d
            if d < dmin:
                dmin = d
                jmin = j

        score += 1.0 if (labels[i] == labels[jmin]) else 0.0
    
    return score / len(test)
```
